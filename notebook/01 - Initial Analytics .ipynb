# Initial Analysis of fetched data


```python
import json
import pandas as pd
import numpy as np
import os
import glob
from pandas.io.json import json_normalize


path = "./trainingdata/3l/"
all_files = glob.glob(os.path.join(path, "*.json")) #make list of paths

dfs = []
for file in all_files:
    df = pd.read_json(file, lines=True)
    dfs.append(df)
    
profiles = pd.concat(dfs).drop_duplicates('profileId')

profiles
```

```python

positions = []
for profile_position in profiles.positions:
    for position in profile_position:
        positions.append(position)

positions = pd.DataFrame(positions)
positions['count'] = 1
positions
```

# Distribution

```python
positions['title'].describe()
```

# Top 10 Positions

```python
counts = positions['title'].value_counts()
counts[:10]
```

# Words in Positions


```python
words = positions['title'].str.split("[\W+\s+]").apply(pd.Series).unstack().replace('', np.nan).dropna().reset_index(0,drop=True)
```

```python
import wordcloud 
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

string = " ".join(words.values)

wc = wordcloud.WordCloud(
    background_color="white", 
    max_words=50, 
    max_font_size=200, 
    random_state=42,
    width=1500,
    height=1000,
    collocations=False,
)
wc.generate(string)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```

# Skills


```python
skills = []
for profile_skill in profiles.skills.dropna():
    for skill in profile_skill:
        skills.append(skill)
        
skills = pd.DataFrame(skills)
```

```python
skills['name'].describe()
```

```python
counts = skills['name'].value_counts()
counts[:15]
```

```python
to_plot = counts

y_pos = np.arange(len(to_plot))

plt.bar(y_pos, to_plot.values, 1/1.5)
#plt.xticks(y_pos, to_plot.index)
plt.show()
```
